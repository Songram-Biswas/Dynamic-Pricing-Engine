```python
# # import pandas as pd
# # import pymongo

# # # Load the core datasets from your local directory
# # items = pd.read_csv('olist_order_items_dataset.csv')
# # orders = pd.read_csv('olist_orders_dataset.csv')
# # products = pd.read_csv('olist_products_dataset.csv')
# # translation = pd.read_csv('product_category_name_translation.csv')
# # customers = pd.read_csv('olist_customers_dataset.csv')
# # sellers = pd.read_csv('olist_sellers_dataset.csv')

# # # Merge the dataframes step-by-step
# # df = pd.merge(items, orders, on='order_id', how='left')
# # df = pd.merge(df, products, on='product_id', how='left')
# # df = pd.merge(df, translation, on='product_category_name', how='left')
# # df = pd.merge(df, customers, on='customer_id', how='left')
# # df = pd.merge(df, sellers, on='seller_id', how='left')

# # # Save the combined dataset to your folder
# # df.to_csv('combined_olist_dataset.csv', index=False)

# # print("File 'combined_olist_dataset.csv' has been created in your folder.")
# # import pandas as pd
# # from deep_translator import GoogleTranslator

# # items = pd.read_csv('olist_order_items_dataset.csv')
# # orders = pd.read_csv('olist_orders_dataset.csv')
# # products = pd.read_csv('olist_products_dataset.csv')
# # translation = pd.read_csv('product_category_name_translation.csv')
# # customers = pd.read_csv('olist_customers_dataset.csv')
# # sellers = pd.read_csv('olist_sellers_dataset.csv')
# # reviews = pd.read_csv('olist_order_reviews_dataset.csv')

# # extra_translations = pd.DataFrame({
# #     'product_category_name': ['pc_gamer', 'portateis_cozinha_e_preparadores_de_alimentos'],
# #     'product_category_name_english': ['pc_gamer', 'kitchen_and_food_preparators_portables']
# # })
# # translation = pd.concat([translation, extra_translations], ignore_index=True)

# # df = pd.merge(items, orders, on='order_id', how='left')
# # df = pd.merge(df, products, on='product_id', how='left')
# # df = pd.merge(df, translation, on='product_category_name', how='left')
# # df = pd.merge(df, customers, on='customer_id', how='left')
# # df = pd.merge(df, sellers, on='seller_id', how='left')
# # df = pd.merge(df, reviews, on='order_id', how='left')

# # translator = GoogleTranslator(source='pt', target='en')

# # def translate_text(text):
# #     if pd.isna(text) or text == '':
# #         return text
# #     try:
# #         return translator.translate(text)
# #     except:
# #         return text

# # # For your demo, translate the first 100 reviews to save time
# # df_sample = df.head(10).copy()
# # df_sample['review_comment_message_en'] = df_sample['review_comment_message'].apply(translate_text)
# # df_sample['review_comment_title_en'] = df_sample['review_comment_title'].apply(translate_text)

# # df_sample.to_csv('combined_olist_dataset_en.csv', index=False)
# import pandas as pd

# items = pd.read_csv('olist_order_items_dataset.csv')
# orders = pd.read_csv('olist_orders_dataset.csv')
# products = pd.read_csv('olist_products_dataset.csv')
# translation = pd.read_csv('product_category_name_translation.csv')
# customers = pd.read_csv('olist_customers_dataset.csv')
# sellers = pd.read_csv('olist_sellers_dataset.csv')

# extra_map = pd.DataFrame({
#     'product_category_name': ['pc_gamer', 'portateis_cozinha_e_preparadores_de_alimentos'],
#     'product_category_name_english': ['pc_gamer', 'kitchen_appliances']
# })
# translation = pd.concat([translation, extra_map], ignore_index=True)

# df = pd.merge(items, orders, on='order_id', how='left')
# df = pd.merge(df, products, on='product_id', how='left')
# df = pd.merge(df, translation, on='product_category_name', how='left')
# df = pd.merge(df, customers, on='customer_id', how='left')
# df = pd.merge(df, sellers, on='seller_id', how='left')

# df['product_category_name_english'] = df['product_category_name_english'].fillna(df['product_category_name'])

# df.to_csv('final_pricing_dataset_mongodb.csv', index=False)
# print("File 'combined_olist_dataset.csv' has been created in your folder.")
import pandas as pd

items = pd.read_csv('olist_order_items_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')
translation = pd.read_csv('product_category_name_translation.csv')
customers = pd.read_csv('olist_customers_dataset.csv')
sellers = pd.read_csv('olist_sellers_dataset.csv')
reviews = pd.read_csv('olist_order_reviews_dataset.csv')
payments = pd.read_csv('olist_order_payments_dataset.csv')

extra_translations = pd.DataFrame({
    'product_category_name': [
        'pc_gamer', 
        'portateis_cozinha_e_preparadores_de_alimentos', 
        'la_cuisine', 
        'cds_dvds_musicais'
    ],
    'product_category_name_english': [
        'pc_gamer', 
        'kitchen_and_food_preparators_portables', 
        'la_cuisine', 
        'cds_dvds_musicals'
    ]
})
translation = pd.concat([translation, extra_translations], ignore_index=True)

df = pd.merge(items, orders, on='order_id', how='left')
df = pd.merge(df, products, on='product_id', how='left')
df = pd.merge(df, translation, on='product_category_name', how='left')
df = pd.merge(df, customers, on='customer_id', how='left')
df = pd.merge(df, sellers, on='seller_id', how='left')
df = pd.merge(df, reviews, on='order_id', how='left')
df = pd.merge(df, payments, on='order_id', how='left')

df['product_category_name_english'] = df['product_category_name_english'].fillna('other')

df.to_csv('final_pricing_dataset_with_reviews.csv', index=False)
```
